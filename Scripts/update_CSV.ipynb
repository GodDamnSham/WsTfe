```python
#combine the data haven same folder number
def combine_data(csv_file):
    data_dict = {}
    with open(csv_file, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader) 
        for row in csv_reader:
            id_publiction = int(row[1])
            images_data = row[2]
            if id_publiction not in data_dict:
                data_dict[id_publiction] = []
            data_dict[id_publiction].append(images_data)
    return data_dict
```

```python
data_dict_csv1 = combine_data("/Users/malik/Downloads/llm.csv")
print(data_dict_csv1)
```

```python
# combine data
#create new colomn , add header in the strava Csv file , add the data from the images in the new updated colomn, place the colomn in csv of strava
def update_csv(csv1, csv2):
    data_dict_csv1 = combine_data(csv1)
    with open(csv2, 'r') as file:
        csv_reader = csv.reader(file)
        updated_colomn = []
        header = next(csv_reader)
        header.extend(["Image_" + time.strftime("%Y%m%d-%H%M")])
        updated_colomn.append(header)
        for colomn_value in csv_reader:
            id_publiction = int(colomn_value[0])
            if id_publiction in data_dict_csv1:
                colomn_value.append({f"Image {i + 1}": data for i, data in enumerate(data_dict_csv1[id_publiction])})
            else:
                colomn_value.append({})
            updated_colomn.append(colomn_value)

    with open(csv2, 'w', newline='') as out_file:
        csv_writer = csv.writer(out_file)
        csv_writer.writerows(updated_colomn)

```
